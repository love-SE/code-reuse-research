```python
import pandas as pd
info_df  =pd.read_excel('./n time modify hit statistics.xlsx')
info_df.info()
```

```python
yes_df = info_df[info_df['hit so nums']!=0]
len(yes_df)
```

```python
yes = (yes_df['modify time cut'].value_counts()/(len(yes_df))).values
yes = [round(i*100,2) for i in yes]
yes
```

```python
no_df = info_df[info_df['hit so nums']==0]
len(no_df)
```

```python
no = (no_df['modify time cut'].value_counts()/len(no_df)).values
no = [round(i*100,2) for i in no]
no
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(len(yes))   
listDate = ["2, 3", "4, 5", "6, 7", "8, 9", ">= 10"]

plt.figure(dpi=200)
total_width, n = 0.8, 2   
width = total_width / n   
x1 = x - width / 2   
x2 = x1 + width   


plt.xlabel("Modification times")   
plt.ylabel("Percentage")  
plt.bar(x1, yes, width=width, label="Reuse to SO",color='green')
plt.bar(x2, no, width=width, label="No reuse to SO", color='orange')
plt.xticks(x, listDate)   

for a, b in zip(x1, yes):
	plt.text(a, b + 0.1, '%.2f' % b, ha='center', va='bottom', fontsize=7)

for a, b in zip(x2, no):
	plt.text(a, b + 0.1, '%.2f' % b, ha='center', va='bottom', fontsize=7)


plt.legend()   
plt.grid()
plt.savefig('./figure 8.jpg',bbox_inches = 'tight')

```
