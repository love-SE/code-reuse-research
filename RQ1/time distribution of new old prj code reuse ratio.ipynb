```python
df2 = pd.read_excel('./01-05 prj-code-reuse-ratio and year.xlsx')
df3 = pd.read_excel('./06-10 prj-code-reuse-ratio and year.xlsx')
df4 = pd.read_excel('./11-15 prj-code-reuse-ratio and year.xlsx')
df5 = pd.read_excel('./16-20 prj-code-reuse-ratio and year.xlsx')
```

```python
sum2 = []
for i in df2.columns[1:]:
    sum2.append(df2[i].sum())
y2 = [(sum2[i+20]/sum2[i])*100 for i in range(20)]


sum3 = []
for i in df3.columns[1:]:
    sum3.append(df3[i].sum())
y3 = []
for i in range(20):
    if sum3[i] == 0:
        y3.append(0)
    else:
       y3.append((sum3[i+20]/sum3[i])*100)


sum4 = []
for i in df4.columns[1:]:
    sum4.append(df4[i].sum())
y4 = []
for i in range(20):
    if sum4[i] == 0:
        y4.append(0)
    else:
       y4.append((sum4[i+20]/sum4[i])*100)



sum5 = []
for i in df5.columns[1:]:
    sum5.append(df5[i].sum())
y5 = []
for i in range(20):
    if sum5[i] == 0:
        y5.append(0)
    else:
       y5.append((sum5[i+20]/sum5[i])*100)

```
