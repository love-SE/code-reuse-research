```python
from datetime import datetime
import time
import os
from tqdm import tqdm
import pandas as pd
import re
import collections
```

```python
df = pd.read_excel('./referred post info(new).xlsx')
df = df[df['referred nums']!=1]
df['set(prj)'] = df['prj'].apply(lambda x: set(re.findall(r'\'(.*?)\'', x)))
df.info()
df.head(2)
```

```python
prjs = ['FirebaseUI-Android','LifeHelper', 'spring-boot', 'OpenHub','xxl-job']

for prj in prjs[:]:
    df1 = df[df['prj'].str.contains(prj)]
    tag_dict = collections.defaultdict(int)
    for str_tags in tqdm(df1['tag']):
        tags = str_tags.strip('<').strip('>').split('><')
        # print(tags)
        for t in tags:
            tag_dict[t]+=1

    tag_dict = sorted(tag_dict.items(), key=lambda x:x[1],reverse=True)
    all_tag_nums = sum([i[1] for i in tag_dict])
    # tag_dict
    tag_list = []
    for info in tag_dict:
        info = list(info)
        info[1] = round(info[1]/all_tag_nums,2)
        tag_list.append([info[0],info[1]])
    print(all_tag_nums)
    print(len(tag_dict))
    print(tag_list[1:11])
```
