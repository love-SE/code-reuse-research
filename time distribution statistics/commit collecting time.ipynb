```python
from tqdm import tqdm
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('./Type_names.txt') as f:
    projects = f.readlines()
    projects = [i.rstrip() for i in projects]
print('prj num{}'.format(len(projects)))
```

```python
# projects=['AarogyaSetu']
year_list = [] # record all commit message path
data_path = 'E:/NewData/'
for project in tqdm(projects):
    project_path = data_path + project
    commits_dir = os.listdir(project_path)
    commits_dir.sort(key=lambda x: int(x))

    for commit in commits_dir:
        commit_path = project_path + '\\' + commit

        with open(commit_path+'\\'+'0.txt', encoding='utf-8') as txt0:
            line = txt0.readline()
            line = txt0.readline()
            line = txt0.readline()
            # line = txt0.readline()

            year = line[-5:-1]
        # bug_flag = False
        # for bug_word in bug_related_words:
        #     if bug_word in line:
        #         bug_flag = True
        #         break
            year_list.append(year)
```

```python
import pandas as pd
x = pd.Series(year_list)
x = x[x!='']
len_x = len(x)
print(x.min(), x.max())
```

```python
x_2000,x_2001,x_2002,x_2003,x_2004,x_2005 = (x=='2000').sum(),(x=='2001').sum(),(x=='2002').sum(),(x=='2003').sum(),(x=='2004').sum(),(x=='2005').sum()
x_2006,x_2007,x_2008 = (x=='2006').sum(),(x=='2007').sum(),(x=='2008').sum()
x_2009,x_2010,x_2011,x_2012,x_2013,x_2014 = (x=='2009').sum(),(x=='2010').sum(),(x=='2011').sum(),(x=='2012').sum(),(x=='2013').sum(),(x=='2014').sum()
x_2015,x_2016,x_2017,x_2018,x_2019,x_2020 = (x=='2015').sum(),(x=='2016').sum(),(x=='2017').sum(),(x=='2018').sum(),(x=='2019').sum(),(x=='2020').sum()
```

### distribution of time from commit SO code reuse pair

```python
# data
x = [i for i in range(1,22)]
y = [x_2000,x_2001,x_2002,x_2003,x_2004,x_2005,
     x_2006,x_2007,x_2008,
    x_2009,x_2010,x_2011,x_2012,x_2013,x_2014,
     x_2015,x_2016,x_2017,x_2018,x_2019,x_2020]
all_year = len_x
plt.figure(dpi=300)
# draw

plt.bar(x,y,align="center",color="#1E90FF",tick_label=[i for i in range(2000,2021)],alpha=1.0)
for a,b,i in zip(x,y,range(len(x))): 
    plt.text(a,b+50,str(y[i]),ha='center',fontsize=5) 
plt.xlabel("Time")
plt.ylabel("Number")
plt.xticks(rotation=45)
plt.grid()
plt.savefig('./distribution of commit occuring time.jpg')
```
